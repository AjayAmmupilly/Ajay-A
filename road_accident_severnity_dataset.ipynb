```python
# AS THERE ARE MANY CATEGORICAL FEATURES WE NEED TO PERFORM FEATURE SELECTION
# USE CHI-SQARE METHOD:Chi-square between each feature and the target and select
# the desired number of features with the best Chi-square scores
```

```python
# STEP1: CONVERT CATEGORICAL FEATURES INTO INTEGERS USING LABEL ENCODING
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# CREATE A NEW DATAFRAME FOR PERFORMING CHI SQUARE ANALYSIS
df1=pd.DataFrame()
```

```python
# ENCODING ALL CATEGORICAL COLUMNS EXCEPT OUTPUT TO NEW DF
for i in cat:
  if i !='Accident_severity':
    df1[i]=le.fit_transform(df[i])
```

```python

# CHECKING THE NEW DF: NOW ALL COLUMNS IN INTEGER
df1.info()
```

```python
# PERFORM CHI SQUARE TEST
from sklearn.feature_selection import chi2
f_p_scores=chi2(df1,df['Accident_severity'])
```

```python

# PRINT F_SCORE AND P_VALUES
f_p_scores
```

```python


# FORM A NEW DF TO EVALUATE THE F_P_SCORES FOR FEATURE SELECTION
f_p_df=pd.DataFrame({'features':df1.columns,'f_score':f_p_scores[0],'p_values':f_p_scores[1]})
f_p_df

```

```python
# WE MUST SELECT FEATURES WITH HIGH F_SCORE AND LOW P_VALUES
# SO SORT WITH P_VALUES IN ASC ORDER
f_p_df.sort_values(by='p_values',ascending=True)
```

```python
# DROP FEATURES: WITH LOW F_SCORE
df2=df.drop(['Owner_of_vehicle','Type_of_vehicle','Road_surface_conditions',
             'Pedestrian_movement','Casualty_severity','Educational_level','Day_of_week','Sex_of_driver',
             'Road_allignment','Sex_of_casualty'],axis=1)
df2.head()
```

```python

df2.shape
```

```python
df2.info()
```

```python
# CHECK FOR MISSING VALUES
df2.isna().sum()
```

```python
# ASSIGNING CATEGORICAL FEATURES TO A VARIABLE
new_cat=[i for i in df2.columns if df2[i].dtype=='O']
print("ALL CATEGORICAL FEATURES IN NEW DF:",new_cat)
```

```python
# COUNT DISTINCT VALUES IN EACH CATEGORIES
for i in new_cat:
  print("DISTINCT VALUES IN",i,"ARE:\n")
  print(df2[i].value_counts())
  print("-"*50)
```

```python
# CONVERTING THE CATEGORICAL FETURES TO INTEGERS
# GET DUMMY
dummy=pd.get_dummies(df2[['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
       'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction',
       'Road_surface_type', 'Light_conditions','Weather_conditions','Type_of_collision',
       'Vehicle_movement', 'Casualty_class', 'Age_band_of_casualty',
       'Cause_of_accident']],drop_first=True)
```

```python
dummy.head(3)

```

```python


# CONCATINATE DUMMY AND DF2
df3=pd.concat([df2,dummy],axis=1)

```

```python
df3

```

```python
# DROP UNWANTED INITAIL COLUMNS
df3.drop(['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience',
       'Area_accident_occured', 'Lanes_or_Medians', 'Types_of_Junction',
       'Road_surface_type', 'Light_conditions', 'Type_of_collision','Weather_conditions',
       'Vehicle_movement', 'Casualty_class', 'Age_band_of_casualty',
       'Cause_of_accident'],axis=1,inplace=True)
```

```python
# FINAL PURE DATASET
df3.head()
```

```python
# CHECK FOR MISSING VALUES
df3.isna().sum()
```

```python
# SEPERATING INPUT X AND OUTPUT Y
x=df3.drop(['Accident_severity'],axis=1)
```

```python
x.head(3)
```

```python
x.shape
```

```python
y=df3['Accident_severity']
```

```python
y.head()
```

```python

# COUNT THE OCCURENCES IN OUTPUT
y.value_counts()
```

```python
# WE CAN SEE THAT THE OUTPUT COUNTS ARE NOT WITHIN THE SAME RANGE
# SO TO IMPROVE MODEL ACCURACY PERFORM UNDERSAMPLING OR OVERSAMPLING

# OVERSAMPLING
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python


# CHECKING THE OUPUT AFTER OVERSAMPLING
y1=pd.DataFrame(yo)
y1.value_counts()


```

```python
# SPLITTING DATA TO TRAIN TEST (TRAIN=70% TEST=30%)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.30,random_state=42)
```

MODEL CREATION


```python

# KNN
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier()
model_knn.fit(x_train,y_train)
y_knn=model_knn.predict(x_test)
y_knn
```

```python
# PERFORMANCE EVALUATION
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
mat=confusion_matrix(y_test,y_knn)
mat
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_knn))


```

```python

score_knn=accuracy_score(y_test,y_knn)
score_knn
```

```python

report=classification_report(y_test,y_knn)
print(report)
```

```python
# NAIVE BAYES
from sklearn.naive_bayes import MultinomialNB
model_nb=MultinomialNB()
model_nb.fit(x_train,y_train)
y_nb=model_nb.predict(x_test)
y_nb
```

```python
# PERFORMANCE EVALUATION
mat_nb=confusion_matrix(y_test,y_nb)
mat_nb
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_nb))
```

```python

score_nb=accuracy_score(y_test,y_nb)
```

```python
report_nb=classification_report(y_test,y_nb)
print(report_nb)
```

```python

# SVM
from sklearn.svm import SVC
model_svm=SVC()
model_svm.fit(x_train,y_train)
y_svm=model_svm.predict(x_test)
y_svm
```

```python
# PERFORMANCE EVALUATION
mat_svm=confusion_matrix(y_test,y_svm)
mat_svm
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_svm))
```

```python
score_svm=accuracy_score(y_test,y_svm)
score_svm
```

```python
report_svm=classification_report(y_test,y_svm)
print(report_svm)
```

```python
# DECISION TREE
from sklearn.tree import DecisionTreeClassifier
model_tree=DecisionTreeClassifier(criterion='entropy')
model_tree.fit(x_train,y_train)
y_tree=model_tree.predict(x_test)
y_tree
```

```python
# PERFORMANCE EVALUATION
mat_tree=confusion_matrix(y_test,y_tree)
mat_tree
```

```python

print(ConfusionMatrixDisplay.from_predictions(y_test,y_tree))
```

```python
score_tree=accuracy_score(y_test,y_tree)
score_tree
```

```python
report_tree=classification_report(y_test,y_tree)
print(report_tree)
```

```python
# RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
model_random=RandomForestClassifier(n_estimators=5,criterion='entropy')
model_random.fit(x_train,y_train)
y_random=model_random.predict(x_test)
```

```python
# PERFORMANCE EVALUATION
mat_random=confusion_matrix(y_test,y_random)
mat_random
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_random))
```

```python
score_random=accuracy_score(y_test,y_random)
score_random
```

```python
report_random=classification_report(y_test,y_random)
print(report_random)
```

```python
# COMPARING ACCURACY SCORES OF VARIOUS CLASSIFICATION ALGORITHMS
algorithms=["knn","naive_bayes","svm","decisiion_tree","random_forest"]
accuracy=[score_knn,score_nb,score_svm,score_tree,score_random]
accuracy_df=pd.DataFrame({'ALGORITHM':algorithms,"ACCURACY":accuracy})
accuracy_df.ACCURACY=accuracy_df.ACCURACY*100
accuracy_df.sort_values(by='ACCURACY',ascending=False,inplace=True)
accuracy_df
```

```python

plt.figure(figsize=(10,5))
sns.barplot(x='ALGORITHM',y='ACCURACY',data=accuracy_df,palette='muted')
font={'family':'serif','color':'black','size':20,'fontweight':'bold'}
plt.title("COMPARING THE ACCURACIES OF THE CLASSIFIACTION ALGORITHMS",fontdict=font)
plt.ylabel("ACCURACY SCORES")
plt.show()
```
